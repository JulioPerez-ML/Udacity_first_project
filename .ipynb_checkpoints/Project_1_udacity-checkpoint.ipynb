#### What's the evolution of average salary in the last years?

```python
fig = plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')

a=plt.scatter(data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'male'].year,
            data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'male'].gross_income_euros,label='Male')
b=plt.scatter(data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'female'].year,
            data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'female'].gross_income_euros, label='Female')
c=plt.scatter(data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'both'].year,
            data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'both'].gross_income_euros, label='Both')

plt.xticks(list((data_nationwide_mean.loc[data_nationwide_mean['gender'] == 'male'].year)))
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('Average Salary in Spain by Gender', fontstyle='oblique', fontsize=15)
plt.legend()

plt.legend( loc='upper left',bbox_to_anchor=(1, 1))

plt.show()
```

#### What about the difference by states?

```python
data_mean_states=data[data['statistic']=='mean']
data_mean_states=data_mean_states[~data_mean_states['state'].isin(['nationwide'])]
data_mean_states.reset_index(drop=True)# we use reset_index to reset the index from 0

table_states_salaries=pd.pivot_table(data_mean_states,index=["state"],values=["gross_income_euros"],
               columns=["gender"])
table_states_salaries
```

```python
table_states_salaries=table_states_salaries['gross_income_euros'].sort_values(by='male',ascending=False).reset_index()

table_states_salaries
```

#### Getting the mean by state

```python
table_states_salaries['mean'] = table_states_salaries[['both', 'male', 'female']].mean(numeric_only=True, axis=1).round(2)
table_states_salaries
```

#### What are the state which has highest salary and lowest salary?

```python
table_first_and_last = pd.concat([table_states_salaries.head(1),table_states_salaries.tail(1)], ignore_index=True)
table_first_and_last
```

```python
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')

n_group = len(table_first_and_last['state'].unique())
index_=np.arange(n_group)
bar_width = 0.25
opacity = 0.8

plt.barh(index_, table_first_and_last['mean'], bar_width, ## using the horizontal bar chart
alpha=opacity,
color='g',
label='Mean Salary')

# plt.xlabel('Euros', fontsize=15)
# plt.ylabel('Euros', fontsize=15)

plt.title('Where is the Highest and Lowest Average Salary?', fontstyle='oblique', fontsize=25)

plt.yticks(index_,table_first_and_last['state'].unique(),rotation=0,fontsize=15)

# using plt.text to place the value labels manually
for i, v in enumerate(table_first_and_last['mean']):
    plt.text(v-5000, i-0.01, str(v)+ '€', color='red', fontweight='bold', fontsize=25)

plt.legend(fontsize=20)
plt.tight_layout()

plt.savefig('picture_1.png', bbox_inches='tight')
plt.show()
```

#### Let's look at the differences among states

## 2. Explain the difference in highest salaries by autonomous region

#### In this section we analyse the highest salaries by region and gender. We must first prepare the subset

```python
data
```

```python
print(data['state'].unique())
print(data['gender'].unique())
print(data['statistic'].unique())
```

```python
data_notNationwide=data[data['state'] != 'nationwide'] # filtering by all all the states and not including 'nationwide'
```

```python
data_notNationwide_upper_and_90= data_notNationwide[data['statistic'].isin(['90th_percentile','upper_quartile'])].reset_index(drop=True)
```

```python
data_notNationwide_upper_and_90.head(50)
```

#### Highest salaries by states

```python
# Order pivot tables https://stackoverflow.com/questions/41119623/pandas-pivot-table-sort-values-by-columns

# we want a pivot table that shows the incomes order descendent by upper_quartile and 90th quartile
table_high_salaries= pd.pivot_table(data_notNationwide_upper_and_90,index=["state","gender"],values=["gross_income_euros"],
               columns=["statistic"])

table_high_salaries.reindex(table_high_salaries['gross_income_euros'].sort_values(by=['upper_quartile','90th_percentile'], ascending=False).index)
```

#### Where are higher salaries by gender?

```python
table_high_salaries_2= pd.pivot_table(data_notNationwide_upper_and_90,index=["state","gender"],values=["gross_income_euros"],
               columns=["statistic"])
#you must order by value. That's the reason you must put table_high_salaries_2['state']. pivot table make the mean of all the
#years
table_high_salaries_2.reindex(table_high_salaries_2['gross_income_euros'].sort_values(by=['gender','upper_quartile'], ascending=False).index)
```

#### A brief look at salaries by state

```python
table_high_salaries_3= pd.pivot_table(data_notNationwide_upper_and_90,index=["state","statistic"],values=["gross_income_euros"],
               columns=["gender"])
table_high_salaries_3
```

#### We want to know the difference between male a female in high salaries and the highest salaries by states

```python
# we calculate the difference and create a new column with the difference value

table_high_salaries_3['diff']=table_high_salaries_3['gross_income_euros']['male']-table_high_salaries_3['gross_income_euros']['female']
```

```python
# we order by statistic value and the difference of salary between male and female

table_high_salaries_3=table_high_salaries_3.reindex(table_high_salaries_3.sort_values(by=['statistic','diff'],ascending=False).index)

table_high_salaries_3
```

#### Plot these differences in a bar chart 

```python

g=pd.DataFrame(data = table_high_salaries_3['diff'])

#to undo the pivot_table indexes use .reset_index()
g=g.sort_values(by='state', ascending=True).reset_index()

#Filtering by each statistic value and creating specific dataframes
diff_90th_percentile=g[g['statistic']=='90th_percentile']
diff_90th_percentile
diff_upper_quartile=g[g['statistic']=='upper_quartile']
diff_upper_quartile

n_groups_ = len(g['state'].unique()) #number of differents categories 

plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')

# getting the numbers of values in X axis
index_=np.arange(n_groups_)

bar_width = 0.35
opacity = 0.8

rects1_ = plt.bar(index_, diff_90th_percentile['diff'], bar_width,
alpha=opacity,
color='b',
label='Extremely High Salary Difference')
#index+bar_width is given to separate the columns so that they do not overlap each other.
rects2_ = plt.bar(index_ + bar_width, diff_upper_quartile['diff'], bar_width,
alpha=opacity,
color='g',
label='High Salaries Difference')

plt.ylabel('Euros', fontsize=15)


plt.title('Difference of Average High Salaries between Male and Female 2008-2017', fontstyle='oblique', fontsize=25)
plt.xticks(index_,diff_upper_quartile['state'].unique(),rotation=75,fontsize=15)
plt.legend(fontsize=15)

plt.tight_layout()
plt.savefig('picture_3.png')
plt.show()
```

### Predicción de salarios 4 próximos años

```python
# # ARIMA example
# from statsmodels.tsa.arima.model import ARIMA
# from random import random
# # contrived dataset
# data_ = [24203.33,
# 25001.05,
# 25479.74,
# 25667.89,
# 25682.05,
# 25675.17,
# 25727.24,
# 25992.76,
# 25924.43, 26391.84]
# # fit model
# model = ARIMA(data_, order=(1, 1, 1))
# model_fit = model.fit()
# # make prediction
# yhat = model_fit.predict(len(data_), len(data_)+5, typ='levels')
# print(yhat)
```

```python

data_nationwide_mean_male = data_nationwide_mean[data_nationwide_mean['gender'].isin(['male'])].sort_values(by='year',ascending=True).reset_index(drop=True)
data_nationwide_mean_female = data_nationwide_mean[data_nationwide_mean['gender'].isin(['female'])] .sort_values(by='year',ascending=True).reset_index(drop=True)
data_nationwide_mean_both = data_nationwide_mean[data_nationwide_mean['gender'].isin(['both'])] .sort_values(by='year',ascending=True).reset_index(drop=True)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model_1 = ARIMA(data_nationwide_mean_male.gross_income_euros,order=(1,1,1))
model_1_fit = model_1.fit()
yhat_1 = model_1_fit.predict(len(data_nationwide_mean_male.gross_income_euros), len(data_nationwide_mean_male.gross_income_euros) + 3)

model_2 = ARIMA(data_nationwide_mean_female.gross_income_euros,order=(1,1,1))
model_2_fit = model_2.fit()
yhat_2 = model_2_fit.predict(len(data_nationwide_mean_female.gross_income_euros), len(data_nationwide_mean_female.gross_income_euros) + 3)

model_3 = ARIMA(data_nationwide_mean_both.gross_income_euros,order=(1,1,1))
model_3_fit = model_3.fit()
yhat_3 = model_3_fit.predict(len(data_nationwide_mean_both.gross_income_euros), len(data_nationwide_mean_both.gross_income_euros) + 3)
```

```python
data_nationwide_mean_male.reset_index(drop=True);
data_nationwide_mean_female.reset_index(drop=True);
data_nationwide_mean_both.reset_index(drop=True);
```

```python
print(yhat_1,yhat_2,yhat_3)
```

```python
more_years = pd.Series([2018,2019,2020,2021])
year = data_nationwide_mean_male.year
year=year.append(more_years).reset_index(drop=True)
p=list(yhat_1.round(2))
p.insert(0,26391.84)

p_2=list(yhat_2.round(2))
p_2.insert(0,20607.85)

p_3=list(yhat_3.round(2))
p_3.insert(0,23646.50)


a=pd.DataFrame({'year':data_nationwide_mean_male.year,'gross_income_euros':data_nationwide_mean_male.gross_income_euros})
prediction=pd.DataFrame({'year': [2017,2018,2019,2020,2021],
                        'predictions': p})

b = pd.DataFrame({'year':data_nationwide_mean_female.year,'gross_income_euros':data_nationwide_mean_female.gross_income_euros})
prediction_2=pd.DataFrame({'year': [2017,2018,2019,2020,2021],
                        'predictions': p_2})

c = pd.DataFrame({'year':data_nationwide_mean_both.year,'gross_income_euros':data_nationwide_mean_both.gross_income_euros})
prediction_3=pd.DataFrame({'year': [2017,2018,2019,2020,2021],
                        'predictions': p_3})
```

```python
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(year[0:len(a.gross_income_euros)], a.gross_income_euros, 'r',label='Male',lw = 4)
plt.plot(year[9:14], prediction.predictions, 'b',linestyle='-.',lw = 8,label='Predictions')

plt.plot(year[0:len(b.gross_income_euros)], b.gross_income_euros, 'g',label='Female',lw = 4)
plt.plot(year[9:14], prediction_2.predictions, 'b',linestyle='-.',lw = 8)

plt.plot(year[0:len(c.gross_income_euros)], c.gross_income_euros, 'y',label='Both',lw = 4)
plt.plot(year[9:14], prediction_3.predictions, 'b',linestyle='-.',lw = 8)

plt.xticks(year)

plt.ylabel('Salary', fontsize=15)
plt.title('Predictions of Salary by Gender in Spain', fontstyle='oblique', fontsize=25)
plt.legend(fontsize=15)

plt.savefig('picture_5.png', bbox_inches='tight')
plt.show()
```
